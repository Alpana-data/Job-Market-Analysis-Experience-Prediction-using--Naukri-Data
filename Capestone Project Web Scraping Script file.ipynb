```python
import bs4
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
import time 
```

```python
driver = webdriver.Edge()
all_data1 = []
for page in range(1, 51): 
    if page == 1:
        url = "https://www.naukri.com/data-science-jobs?src=gnbjobs_homepage_srch"
    else:
        url = f"https://www.naukri.com/data-science-jobs-{page}?src=gnbjobs_homepage_srch"
    driver.get(url)
    
    time.sleep(5)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    jobs = soup.find_all("div", class_="srp-jobtuple-wrapper")
   
    print(f"Page {page}, Jobs found:", len(jobs))
    for job in jobs:
        try:
            Titles = job.find("a", class_="title").text.strip()
        except:
            Titles = ""
        try: 
            Experience = job.find("span", class_="expwdth").text.strip()
        except: 
            Experience = ""
        try: 
            Location= job.find("span", class_="locWdth").text.strip()
        except: 
            Location = ""
        try: 
            Company_name= job.find("a", class_="comp-name mw-25").text.strip()
        except: 
            Company_name = ""
        try: 
            Job_Educational_requirements= job.find("span", class_="job-desc ni-job-tuple-icon ni-job-tuple-icon-srp-description").text.strip()
        except: 
            Job_Educational_requirements = ""
        try: 
            Skills_required = job.find("ul", class_="tags-gt").text.strip()
        except: 
            Skills_required= ""
        try: 
            Total_reviews = job.find("a", class_="review ver-line").text.strip()
        except: 
            Total_reviews= ""
        try: 
            Rating = job.find("span", class_="main-2").text.strip()
        except: 
            Rating= ""
        try:
            job_post_day = job.find("span", class_="job-post-day").text.strip()
        except:
            job_post_day=""
        all_data1.append([Titles, Experience, Location, Company_name, Job_Educational_requirements,Skills_required, Total_reviews, Rating, job_post_day])
```

```python
df1 = pd.DataFrame(all_data1, columns=["Titles", "Experience", "Location", "Company_name", "Job_Educational_requirements", 
"Skills_required", "Total_reviews", "Rating","job_post_day"])
print(len(all_data1))
df1.to_csv("naukri_jobslist1.csv", index=False)
print(f"Saved till page {page}")
```
